```python
import pandas as pd

# Load datasets
mat = pd.read_csv("student-mat.csv", sep=';')
por = pd.read_csv("student-por.csv", sep=';')

# Columns to use for merging (identifiers)
merge_keys = [
    "school", "sex", "age", "address", "famsize", "Pstatus",
    "Medu", "Fedu", "Mjob", "Fjob", "reason", "guardian"
]

# Merge the datasets â€” students in both math & Portuguese will appear once
merged = pd.merge(mat, por, on=merge_keys, suffixes=('_mat', '_por'))

# Save to file
merged.to_csv("student-merged.csv", index=False)

print("Merged shape:", merged.shape)
print("Example rows:")
print(merged.head())
```
